#### 8. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userID,ViewCount,CommentCount

```python
new_users = users[['userId', 'Reputation', 'Views', 'UpVotes', 'DownVotes']]
new_posts = posts[['postId', 'Score', 'userId', 'ViewCount', 'CommentCount']]
print(new_users.head())
print(new_posts.head())
```

#### 8. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
merge_df = new_users.merge(new_posts, left_on='userId', right_on='userId')
merge_df.head()
```

#### 11. Adjust the data types in order to avoid future issues. Which ones should be changed? 

```python
merge_df.dtypes
```

```python

merge_df['ViewCount'] = merge_df['ViewCount'].astype('int64')

#merge_df['userId'] = merge_df['userId'].astype('object')
merge_df.head()
```

```python
stats = merge_df.describe().transpose()
stats['IQR'] = stats['75%'] - stats['25%']
stats
```

```python
outliers = pd.DataFrame(columns=merge_df.columns)

for col in stats.index:
    iqr = stats.at[col,'IQR']
    cutoff = iqr * 3
    lower = stats.at[col,'25%'] - cutoff
    upper = stats.at[col,'75%'] + cutoff
    results = merge_df[(merge_df[col] < lower) | 
                   (merge_df[col] > upper)].copy()
    results['Outlier'] = col
    outliers = outliers.append(results)
```
