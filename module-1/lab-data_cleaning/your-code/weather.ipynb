## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import numpy as np
import pandas as pd
```

```python
folder = "/Users/admin/Documents/data-labs/module-1/lab-data_cleaning/"
weather = pd.read_csv(folder+'weather-raw.csv')
```

```python
weather
```

```python

melted = weather.melt(id_vars=['id', 'year', 'month', 'element'], 
                    var_name='Day', value_name='Values')

melted.head()
```

```python
melted = melted.dropna(subset=['Values']) 
melted
```

```python
#melted = melted.drop('Date', axis=1)
```

```python
melted['Date'] = melted['year'].map(str) +'/'+ melted['month'].map(str) +'/'+ melted['Day'].str.replace('d', '')
```

```python
melted
```

```python
melted = melted.drop(['id','year', 'month', 'Day'], axis=1)
```

```python
melted
```

```python
result = melted.pivot(index='Date', columns='element', values='Values')
```
