```python
#data[data['Date'].str.contains('[bB]efore\d\d-\w\w\w-\d{4}', na = False, regex= True)]
import re


data.loc[data['Date'].str.contains('[bB]efore\d\d-\w\w\w-\d{4}', na = False, regex= True), 'Date']

for elem in data.loc[data['Date'].str.contains('[bB]efore\d\d-\w\w\w-\d{4}', na = False, regex= True), 'Date'] :
    #print(elem)
    #print(re.findall('\d\d-\w\w\w-\d{4}', elem))
    data.loc[data['Date'] == elem, 'Date'] = re.findall('\d\d-\w\w\w-\d{4}', elem)


    
data.loc[data['Date'].str.contains('[bB]efore\d\d-\w\w\w-\d{4}', na = False, regex= True), 'Date']
```

```python
#print(data.loc[data['Date'].str.contains('[bB]efore', na = False, regex= True), 'Date'])
data['Date'] = data['Date'].str.replace('Nodate,', '')
data['Date'] = data['Date'].str.replace(',', '')

for elem in data.loc[data['Date'].str.contains('Before\d{4}', na = False, regex= True), 'Date'] :
    #print(elem)
    #print(int(re.findall('\d{4}', elem)[0])-1)
    data.loc[data['Date'] == elem, 'Date'] = elem.replace('Before', '01-Jan-')
    
    currentyear = int(re.findall('\d{4}', elem)[0])
    newyear = int(re.findall('\d{4}', elem)[0])-1
    
    data.loc[data['Date'] == elem, 'Date'] = elem.replace(str(currentyear),  str(newyear))
    

```
