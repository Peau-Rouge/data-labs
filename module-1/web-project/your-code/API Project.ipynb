```python

import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
url = 'https://age-of-empires-2-api.herokuapp.com/api/v1'
endpoint = '/structures'
response = requests.get(url+endpoint, headers={"user-agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36"})

results = response.json()
data = pd.DataFrame(results)
#data.head(10)

df = json_normalize(data.structures)

df
```

```python
df.groupby(['age']).sum().build_time.sort_values(ascending=False)
```

```python
df.groupby(['age']).mean().hit_points.sort_values(ascending=False)
```

```python
df.groupby(['age','name']).mean().hit_points
```

```python
df.groupby(['age','name']).mean().line_of_sight
```

```python
df.groupby(['age','name']).mean()['cost.Wood']
```
