```python
import pandas as pd
import plotly.plotly as py
import cufflinks as cf
from ipywidgets import interact

cf.go_offline()
```

```python
data = pd.read_excel('../data/Online Retail.xlsx')
data.head()
```

## 1. Create an interactive bar chart showing total quantity and revenue by country (excluding United Kingdom) for the month of April 2011.

```python
df = data[data['Country'] != 'United Kingdom']
#df.head(10)
```

```python
df.dtypes
df['month_year'] = pd.to_datetime(df['InvoiceDate']).dt.to_period('M')
df['day_month_year'] = pd.to_datetime(df['InvoiceDate']).dt.to_period('D')
df.head()
#April2011.head()
                     
```

```python
April2011 = df[df['month_year']=='2011-04']

april_groupby = April2011.groupby('Country', as_index=False).agg({'Quantity':'sum', 'Revenue':'sum'})
april_groupby
```

```python

april_groupby.iplot(kind='bar', x ='Country', xTitle='Country',
           yTitle='Total Quantity and Revenue', title='Total Quantity and Revenue by Country for April2011')
```

## 2. Create an interactive line chart showing quantity and revenue sold to France between January 1st and May 31st 2011.

```python
df['year'] = pd.DatetimeIndex(df['InvoiceDate']).year
df['month'] = pd.DatetimeIndex(df['InvoiceDate']).month
df['day'] = pd.DatetimeIndex(df['InvoiceDate']).day

df.head()
```

```python
data_france = df[df['Country'] == 'France']
data_france2011 = data_france[(data_france['year']==2011)]
jan_may = data_france2011[(data_france2011['month'] >= 1) & (data_france2011['month'] <= 5)]
jan_may.head()
```

```python
france_groupby = jan_may.groupby('day_month_year', as_index=False).agg({'Quantity':'sum', 'Revenue':'sum'})

france_groupby.iplot(kind='line', x='day_month_year', xTitle='time',
           yTitle='Quantity and Revenue', title='Quantity and Revenue in France')
```

## 3. Create an interactive scatter plot showing the relationship between average quantity (x-axis) and average unit price (y-axis) for the product PARTY BUNTING with the plot points color-coded by country (categories).

```python
df = data[data['Description'] == 'PARTY BUNTING']
df.head(10)

avg = df.groupby('Country', as_index=False).agg({'Quantity':'mean', 'UnitPrice':'mean'})
avg
```

```python
avg.iplot(x='Quantity', y='UnitPrice', categories='Country',
           xTitle='Quantity', yTitle='UnitPrice',
           title='relationship between average quantity and average unit price for the product PARTY BUNTING')
```

## 4. Create a set of interactive histograms showing the distributions of quantity per invoice for the following countries: EIRE, Germany, France, and Netherlands.

```python
df = data[(data['Country'] == 'EIRE') | (data['Country'] == 'Germany') | (data['Country'] == 'France') | (data['Country'] == 'Netherlands')]
df.head(10)

qty = df.pivot_table(values='Quantity', columns='Country', index='InvoiceNo', aggfunc='sum')

qty.head(10)

```

```python
qty.iplot(kind='hist',  xTitle='InvoiceNo', subplots=True,
yTitle='Quantity', title='Distribution of Quantity for 4 countries')
```
